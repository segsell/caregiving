```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 20, 400)
```

```python
# Define epsilon
eps = 0.1
scale = 0.2

1 / eps
```

```python


# Compute reciprocal-like functions
y_inv_sqrt_eps = 1 / (np.sqrt(scale * x) + eps)
y_inv_log_eps = 1 / np.log(scale * x + 1 + eps)

# Plot
plt.figure(figsize=(8,6))
plt.plot(x, y_inv_sqrt_eps, label=rf'$1/\sqrt{{x+{eps}}}$')
plt.plot(x, y_inv_log_eps, label=rf'$1/\log(x+1+{eps})$')

plt.ylim(0, 12)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Reciprocal-like functions with epsilon = 0.1")
plt.legend()
plt.grid(True)
```
